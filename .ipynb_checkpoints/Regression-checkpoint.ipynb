```python
import numpy as np
from gensim.models import Word2Vec
import matplotlib.pyplot as plt
from sklearn import linear_model
import time
```

```python
wv = np.load('./model/wv.npy')
vocab = []
with open('./data/common_word_list.txt', encoding = 'utf-8') as f:
    for line in f.readlines():
        vocab.append(line.strip())
print(vocab[:10])
print(wv[:10])
```

```python
print(len(vocab), wv.shape)
```

```python
profs = []
with open('./class/profs.txt', encoding = 'utf-8') as f:
    for word in f.readlines():
        word = word.strip()
        profs.append(word)
```

```python
print(profs, len(profs))
```

```python
print(wv[vocab.index('教师')][:10])
```

```python
label = np.zeros(len(vocab))
cnt = 0
for word in profs:
    if word not in vocab:
        print(word)
        cnt+=1
    else:
        i = vocab.index(word)
        label[i] = 1
```

```python
r_model = linear_model.LinearRegression()
r_model.fit(wv, label)
display(r_model.intercept_) 
display(r_model.coef_)
```

```python
r_model.coef_.shape
```

```python
weights = r_model.coef_
bias = r_model.intercept_
```

```python
file_name = f'weight_{time.strftime("%y%m%d_%H%M", time.localtime())}.png'
x = np.arange(0, 256, 1)
plt.figure(figsize=(30, 5))
plt.bar(x, r_model.coef_)
plt.savefig(f'./pics/{file_name}')
plt.show()
```

```python
# 以下用一个画布多个子图重画

x = np.arange(0, 256, 1)
plt.figure(figsize=(30, 5))
plt.bar(x, r_model.coef_*wv[vocab.index('教师')])
# plt.savefig('weights_1.png')
plt.show()
```
